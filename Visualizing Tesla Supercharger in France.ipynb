***Let's get started!***

```python
df = pd.read_excel('VisualizingTeslaSupercharger.xlsx')
```

```python
lat = df['Ylatitude']
lng = df['Xlongitude']
station = df['ad_station']
```

```python
map = folium.Map(location=[48.754865,2.373034], zoom_start=5)
market_cluster = MarkerCluster().add_to(map)
for lat, lng, station in zip(lat, lng, station):
    folium.Marker(location=[lat, lng], popup=str(station), icon=folium.Icon(color='green')).add_to(market_cluster)
```

We loop trough the data to get latitude, longitude and station name then we merge using zip and create mark within loop and cluster it using MarektCluster()
